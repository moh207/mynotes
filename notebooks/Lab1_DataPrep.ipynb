# Most No of cancelled orders 

```python
df['Country'] = df['Country'].apply(lambda x: 'United Kingdom' if (x == "UK") or (x == "uk") else x)
df['Country'].unique()
```

```python
dfcanc = df[(df['Canceled'] == 1)].groupby(['Country']).InvoiceNo.count().reset_index(name='No of canceled orders').sort_values('No of canceled orders',ascending=False)
dfcanc.head()
```

```python
dftotal = df.groupby(['Country']).InvoiceNo.count().reset_index(name='Total No of orders').sort_values('Total No of orders',ascending=False)
dftotal.head()
```

```python
merged_df = dfcanc.merge(dftotal, how='inner', left_on=["Country"], right_on=["Country"])
merged_df.head()
```

```python
merged_df['Cancel Ratio'] = merged_df['No of canceled orders'] / merged_df['Total No of orders']
merged_df.head()
```

# USA has the biggest Cancellation percentage with about 41% of orders being cancelled, which is almost half of the ordered items being cancelled

```python
merged_df.sort_values('Cancel Ratio',ascending=False)
```
